# Statistics on detection

## General

```python
df_detections
```

```python
detections_per_shark = df_detections.groupby('Shark').size().reset_index(name='Total Detections')
detections_per_shark
```

```python
detections_per_shark['Total Detections'].sum()
```

```python
Bull_mask = detections_per_shark['Shark'].apply( lambda s: True if (s.startswith('B')) else False )
Tiger_mask = ~Bull_mask

# Split the dataframe into daytime and nighttime dataframes
Bull_detections = detections_per_shark[Bull_mask]
Tiger_detections = detections_per_shark[Tiger_mask]
```

## Bulls & Tigers

```python
print('Bull:',Bull_detections['Total Detections'].describe(), Bull_detections['Total Detections'].sum())
print('\n')
print('Tiger', Tiger_detections['Total Detections'].describe(), Tiger_detections['Total Detections'].sum())
```

```python
Bull_F_mask = Bull_detections['Shark'].apply( lambda s: True if (s.endswith('F')) else False )
Bull_M_mask = ~Bull_F_mask

# Split the dataframe into daytime and nighttime dataframes
Bull_F_detections = Bull_detections[Bull_F_mask]
Bull_M_detections = Bull_detections[Bull_M_mask]
```

```python
Tiger_F_mask = Tiger_detections['Shark'].apply( lambda s: True if (s.endswith('F')) else False )
Tiger_M_mask = ~Tiger_F_mask

# Split the dataframe into daytime and nighttime dataframes
Tiger_F_detections = Tiger_detections[Tiger_F_mask]
Tiger_M_detections = Tiger_detections[Tiger_M_mask]
```

```python
print('Bull Females:',Bull_F_detections['Total Detections'].describe(), Bull_F_detections['Total Detections'].sum())
print('\n')
print('Bull Males:', Bull_M_detections['Total Detections'].describe(), Bull_M_detections['Total Detections'].sum())
```

```python
print('Tiger Females:',Tiger_F_detections['Total Detections'].describe(), Tiger_F_detections['Total Detections'].sum())
print('\n')
print('Tiger Males:', Tiger_M_detections['Total Detections'].describe(), Tiger_M_detections['Total Detections'].sum())
```
